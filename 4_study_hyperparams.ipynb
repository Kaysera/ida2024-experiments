```python
from os.path import join, isfile, isdir
from os import listdir
import pandas as pd
```

```python
my_path = '../results/'
db = 'wine'
method = 'flare'
onlyfiles = [f for f in listdir(my_path) if isfile(join(my_path, f)) and db in f and method in f]
```

```python
res = []
for f in onlyfiles:
    with open(join(my_path, f), 'r') as file:
        params = f[:-4].split('_')
        db = params[2]
        method = params[3]
        population_size = params[4]
        size_pressure = params[5]
        kappa = params[6]
        epsilon = params[7]
        best_flocal_size, best_flocal_score, flocal_score = [float(i) for i in file.read()[1:-1].split(', ')]
        res.append([best_flocal_size, best_flocal_score, flocal_score, db, method, population_size, size_pressure, kappa, epsilon])
```

```python
df = pd.DataFrame(res, columns=['best_flocal_size', 'best_flocal_score', 'flocal_score', 'db', 'method', 'population_size', 'size_pressure', 'kappa', 'epsilon'])
```

```python
df.sort_values(by=['best_flocal_score', 'best_flocal_size', 'population_size'], ascending=[False, True, True])
```
